```python


iris=sns.load_dataset('iris')
```

```python


sns.heatmap(data=iris.corr(),annot=True,cmap="icefire")

```

```python


sns.histplot(data=iris,x="sepal_length",hue="species",kde=True,palette="rocket")
plt.title("Sepal_length")

```

```python


sns.histplot(data=iris,x="sepal_width",hue="species",kde=True,palette="icefire")
plt.title("Sepal_Width")

```

```python
sns.histplot(data=iris,x="petal_length",hue="species",kde=True,palette="coolwarm")
plt.title("Petal Length")
```

```python


sns.histplot(data=iris,x="petal_width",hue="species",kde=True,palette="rocket")
plt.title("Petal Width")

```

```python
sns.jointplot(data=iris,x="petal_width",y="petal_length",color="thistle")
```

```python


plt.pie(x=[50,50,50],labels=["setosa","versicolor","virginica"],autopct="%1.1f%%")
plt.title("Distribution")

```

```python
sns.scatterplot(data=iris,x="sepal_length",y="petal_width")
plt.title("Sepal Length vs Petal Width")
```

```python


sns.lineplot(data=iris,x=iris.index,y="sepal_length")
plt.xlabel("Index")

```

```python
sns.lineplot(data=iris,x=iris.index,y="petal_length")
plt.xlabel("Index")
```

```python


sns.lineplot(data=iris,x=iris.index,y="sepal_width")
plt.xlabel("Index")

```
